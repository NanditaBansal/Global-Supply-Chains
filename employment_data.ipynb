```python
import pandas as pd

# Load the data from the 'employment.csv' file
data = pd.read_csv('employment.csv')

# Define the custom order for the 'ACTIVITY' column
custom_order = [
    'A01_02', 'A03', 'B05_06', 'B07_08', 'B09', 'C10T12', 'C13T15', 'C16', 'C17_18',
    'C19', 'C20', 'C21', 'C22', 'C23', 'C24', 'C25', 'C26', 'C27', 'C28', 'C29', 'C30',
    'C31T33', 'D', 'E', 'F', 'G', 'H49', 'H50', 'H51', 'H52', 'H53', 'I', 'J58T60', 'J61',
    'J62_63', 'K', 'L', 'M', 'N', 'O', 'P', 'Q', 'R', 'S', 'A', 'B', 'C', 'C16T18', 'C19T23',
    'C20_21', 'C24_25', 'C26_27', 'C29_30', 'D_E', 'GTN', 'GTI', 'H', 'J', 'M_N', 'OTT',
    'OTQ', 'RTT', 'R_S', 'BTE', 'GTT', 'JTN', 'INFO'
]

# Check if the 'ACTIVITY' column exists in the dataframe
if 'ACTIVITY' in data.columns:
    # Ensure all activity codes are consistent (e.g., uppercase)
    data['ACTIVITY'] = data['ACTIVITY'].str.upper()

    # Convert the 'ACTIVITY' column to a categorical type with the specified order
    data['ACTIVITY'] = pd.Categorical(data['ACTIVITY'], categories=custom_order, ordered=True)

    # Sort by the 'ACTIVITY' column according to the custom order
    data_sorted = data.sort_values(by='ACTIVITY')

    # Save the sorted data to a new CSV file
    data_sorted.to_csv('employment_sorted.csv', index=False)

```
