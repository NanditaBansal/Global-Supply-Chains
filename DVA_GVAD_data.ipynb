```python
def calculate_activity_share(df):
    # Group by TIME_PERIOD, REF_AREA, and ACTIVITY
    grouped = df.groupby(['TIME_PERIOD', 'REF_AREA'])

    # Define a lambda function to calculate the share for each activity
    def get_share(group):
        total_value = group[group['ACTIVITY'] == '_T']['OBS_VALUE'].values[0]  # Get the total value for each group
        group['SHARE'] = group['OBS_VALUE'] / total_value  # Calculate the share for each activity
        return group

    # Apply the function to each group
    result_df = grouped.apply(get_share)

    # Filter out the total rows (_T) as we only need the share for individual activities
    result_df = result_df[result_df['ACTIVITY'] != '_T']

    # Select relevant columns for the output
    output_df = result_df[['ACTIVITY', 'TIME_PERIOD', 'REF_AREA', 'SHARE']]

    return output_df
```

```python
def multiply_share_with_obs_value(df1, df2):
    # Merge the dataframes on 'TIME_PERIOD' and 'REF_AREA'
    merged_df = pd.merge(df1, df2, on=['TIME_PERIOD', 'REF_AREA'], how='left')

    # Multiply 'OBS_VALUE' from df2 with 'SHARE' from df1 for each row
    merged_df['RESULT'] = merged_df['OBS_VALUE'] * merged_df['SHARE']

    return merged_df
```

```python
def calculate_result_ratio(df):
    # Filter the DataFrame for IND and USA
    df_ind = df[df['REF_AREA'] == 'IND']
    df_usa = df[df['REF_AREA'] == 'USA']

    # Merge IND and USA data on TIME_PERIOD and ACTIVITY_x
    merged_df = pd.merge(
        df_ind, df_usa,
        on=['TIME_PERIOD', 'ACTIVITY_x'],
        suffixes=('_IND', '_USA')
    )

    # Calculate the ratio
    merged_df['RESULT_RATIO'] = merged_df['RESULT_IND'] / merged_df['RESULT_USA']

    # Select relevant columns and drop REF_AREA
    result_df = merged_df[['TIME_PERIOD', 'ACTIVITY_x', 'RESULT_RATIO']]

    return result_df
```

```python
def calculate_result_ratio_rate(df1, df2):
    # Merge df1 and df2 based on TIME_PERIOD and YEAR
    merged_df = pd.merge(df1, df2, left_on='TIME_PERIOD', right_on='YEAR', how='left')

    # Calculate RESULT_RATIO_RATE
    merged_df['RESULT_RATIO_RATE'] = merged_df['RESULT_RATIO'] * merged_df['RATE']

    # Select and reorder columns
    output_df = merged_df[['TIME_PERIOD', 'ACTIVITY_x', 'RESULT_RATIO', 'RESULT_RATIO_RATE', 'RATE']]

    return output_df
```
